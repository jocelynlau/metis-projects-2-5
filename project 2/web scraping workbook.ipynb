```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

import chromedriver_binary
```

```python
# main_url = 'https://en.wikipedia.org/wiki/Lists_of_hospitals_in_the_United_States'
main_url = 'https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States'
```

```python
driver = webdriver.Chrome()
driver.get(main_url)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
tables = soup.find_all('table')
rows = [row for row in tables[0].find_all('tr')]
```

```python
for row in rows[2]:
    driver.find_element_by_link_text("/wiki/Alabama").click()
```

```python
rows[2].find('th')
```

```python
# name_selector = '/wiki/List_of_hospitals_in_Alabama'
# driver.find_element_by_xpath(name_selector).click()
# current_url = driver.current_url
state_list = ['Alabama']

for state in state_list:
    #click on list
    driver.find_element_by_link_text(state).click()
    #pull table info into list using beautiful soup
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    tables = soup.find_all('table')
    rows = [row for row in tables[0].find_all('tr')]

    name = []
    loc = []
    bed_count = []
    trauma_level = []
    notes = []
    st = []

    for row in range(1,len(rows)):
        name.append(rows[row].find_all('td')[0].text)
        loc.append(rows[row].find_all('td')[1].text)
        bed_count.append(rows[row].find_all('td')[2].text)
        trauma_level.append(rows[row].find_all('td')[3].text)
        notes.append(rows[row].find_all('td')[4].text)
        st.append(str(state))

    tuples = list(zip(name,loc,bed_count,trauma_level,notes,state))

    df = pd.DataFrame(tuples, columns = ['Hospital', 'Location','Bed Count','Trauma Level','Notes','State'])
    #return to main page
    driver.back()
```

```python
driver.close()
```
